# Preprocessing

```python
# STARWARS
sw
```

1. Casefolding
2. Trim/strip, remove multiple spaces
3. Drop empty strings
4. Duplicates
5. Append Endline token

```python
def clean(df_):
    df = df_.copy()
    # 0. remove "\n
    df.dialogue = df.dialogue.str.replace('"\\n','',regex=True)
    # 1. Casefolding
    df.dialogue = df.dialogue.str.lower()
    # 2. Trim/strip, remove multiple spaces
    df.dialogue = df.dialogue.str.strip()
    df.dialogue = df.dialogue.str.replace(r'\s+',' ',regex=True)
    # 3. Drop empty strings
    df = df.loc[df.dialogue.str.len() > 0]
    # 4. Remove duplicates
    df = df.drop_duplicates(subset='dialogue')
    # 5. Append endline token [EOL] in the end of each dialogue
    df.dialogue = df.dialogue.apply(lambda x: x + ' [EOL]')
    return df
```

```python
def create_next_line(df_):
    df = df_.copy()
    df['next_line'] = df.dialogue.shift(-1).str.split().str[0]
    return df
```

```python
sw4 = create_next_line(sw4)
sw5 = create_next_line(sw5)
sw6 = create_next_line(sw6)
```
